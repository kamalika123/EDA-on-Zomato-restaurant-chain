```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('zomato.csv')
```

```python
df.info()
```

```python
df = df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
#dropping duplicates
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if (value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value=value[0]
        return float(value)

df['rate'] = df['rate'].apply(handlerate)
df['rate'].sample(7)
```

Filling Null values in Rate with Mean

```python
df['rate'].isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
```

```python
df['rate'].isnull().sum()
```

```python
df.info()
```

Dropping null values of location and cuisine

```python
df.dropna(inplace = True)
```

```python
df.rename(columns={'approx_cost(for two people)':'cost2plates','listed_in(type)':'type'},inplace=True)
```

```python
df.columns
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
df = df.drop(['listed_in(city)'],axis=1)
```

```python
df.columns
```

```python
df['cost2plates'].unique()
```

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
    
df['cost2plates'] = df['cost2plates'].apply(handlecomma)
df['cost2plates'].sample(7)
```

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_less_than_1000 = rest_types[rest_types<1000]
rest_types_less_than_1000
```

Making rest_type having frequency less than 1000 as others

```python
def handle_rest_type(value):
    if(value in rest_types_less_than_1000):
        return 'others'
    else:
        return value
        
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
location = df['location'].value_counts(ascending  = False)
```

```python
location
```

```python
loc_type_less_than_300 = location[location<300]
loc_type_less_than_300
```

```python
def handle_location(value):
    if(value in loc_type_less_than_300):
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handle_location)
    
```

```python
df['location'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending  = False)
```

```python
cuisines_lessthan100 = cuisines[cuisines<100]
```

```python
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
